```python
import random as rm
import matplotlib.pyplot as plt
import numpy as np

a=0
b=1
m=2



def f(x):
  return 2*x

def ace(n):
  lista =[]
  for i in range(n):

    r1 = rm.random()
    r2 = rm.random()
    x = a+(b-a)*r1
    if r2 <= f(x)/m:
      lista.append(x)

  return lista


c = ace(1000)
print(c)
```

```python
plt.hist(ace(1000), bins = 50)
plt.show()
```

```python

print(np.var(c))
print(np.mean(c))
print(np.std(c))
```
