```python
import random as rm
import matplotlib.pyplot as plt
import numpy as np
import math as mt

#PASO 1: VARIABLES
m1 = 1.5
var1 = 0.0016
de1 = 0.04

m2 = 1.48
var2 = 0.0009
de2 = 0.03

def ace(n):
  exito = 0
  for i in range(n):
    x1 = np.random.normal(m1,de1)
    x2 = np.random.normal(m2,de2)
    if x2 >= x1:
      exito += 1
  p = exito/n
  return p

def ace2(M):
  lista = []
  for i in range(M):
    e = ace(10)
    lista.append(e)
  return lista

probabilidad = ace(100000)
print("Probabilidad estimada de que no embonen: ", probabilidad)

m = ace2(500)
print("Probabilidades de cada experimento: " , m)
plt.plot(m)
```

```python
p = probabilidad  # ejemplo estimado
E = 0.001
Z = 1.96

n = (Z * np.sqrt(p * (1 - p)) / E) ** 2
print(f"Nmero mnimo de simulaciones necesarias: {int(np.ceil(n))}")
```

```python
plt.hist(m, bins=30, color='skyblue')
plt.title("Distribucin de probabilidades estimadas en 500 experimentos")
plt.xlabel("Probabilidad de que no embonen")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```
