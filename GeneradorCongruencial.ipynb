```python
import matplotlib.pyplot as plt
import tabulate as tb
import numpy as np

#---------------------------------------------------------#
def Generador(a, c, m, x0):
    return (a * x0 + c) % m
#---------------------------------------------------------#
a = 13
c = 9
m = 128
x0 = 7  # Semilla inicial
#---------------------------------------------------------#

# Generar 10000 números
n_muestras = 10000
muestra = []
valores_vistos = {} # Revisor de valores

for i in range(n_muestras):
    nuevo = Generador(a, c, m, x0)
    muestra.append(nuevo / m)
    x0 = nuevo  # Actualizar la semilla

    if nuevo in valores_vistos:
        print(f"El num. {nuevo} se repite en las iteraciones {valores_vistos[nuevo]} y {i}.")
        print(f"El ciclo de vida es de {i-1} ")
        break
    valores_vistos[nuevo] = i  # Guarda la celda donde esta el num. repetido

# Verificar si hay repetidos en la muestra completa
if len(set(muestra)) == len(muestra):
    print("No hay Numeros repetidos")
    print(" ")
else:
    print("El ciclo no es de 10k, hay un num. repetido.")


#---------------------------------------------------------#
mu = np.mean(muestra)  # Media
sigma2 = np.var(muestra)  # Varianza
print(f" La media es de: {mu}")
print(f" La varianza es de: {sigma2}")

#---------------------------------------------------------#

# Histograma
plt.hist(muestra, bins=20, edgecolor='black', alpha=0.7)
plt.axvline(mu, color='red', linestyle='dashed', linewidth=2, label=f'$\mu$ = {mu:.5f}')
plt.xlabel('Valores Normalizados')
plt.ylabel('Frecuencia')
plt.title('Histograma de 10000 números')
plt.legend()
plt.show()
```
