```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
```

01 - Definição de funções

```python
def g(x):
  return np.sqrt(np.arctan(x))
```

```python
g(1)

a = 1
b = 1
```

02 - Execução

```python
aciertos = 0
fallos = 0
N = 1500
```

```python
X = np.linspace(0,1,100)
Y = g(X)
plt.plot(X,Y, color = 'black')
for _ in range(N):  #Generador de bucle para los puntos
  x = rd.random()
  y = rd.random()
  if y <= g(x):
    plt.scatter(x,y, color = 'red', s=10)
    aciertos += 1 #Estan debajo de la funcion
  else:
    plt.scatter(x,y, color = 'blue', s=10)
    fallos += 1 #Estan arriba de la funcion

print('Aciertos: ', aciertos)
print('Fallos: ', fallos)
I = aciertos/N
print('Valor de la integral: ', I)
plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('Gráfico da função g(x)')
plt.grid()
plt.show()
```
