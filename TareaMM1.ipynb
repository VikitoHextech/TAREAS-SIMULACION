# Línea de espera con un servidor
Usar el código que hicimos en clase para estimar las probabilidades de que haya 1, 2, 3, ... clientes en el sistema (esto ya se esbozo en clase).

Librerias:

```python
import numpy
import matplotlib.pyplot as plt
from random import expovariate
from statistics import mean
from math import inf
```

Parametros y Variables:

```python
lamda = 4.0
mu1 = 6.0  # Servido 1
mu2 = 5.0  # Servidor 2
Num_Pkst = 100000
count = 0
t = 0

n1 = 0 # Clientes servidor 1
n2 = 0 # Clientes servidor 2
tll = expovariate(lamda) # Próxima llegada
ts1 = inf
ts2 = inf

t_evento_anterior = 0.0

tll_Data = [] # llegada al sistema
ts_Data = [] # salida del sistema
r_Data = []
area = []
tiempo_en_estado = {}

```

Simulacion: (simulação para los cuates)

```python
while count < Num_Pkst:

    if tll <= ts1 and tll <= ts2:
        t = tll
        tll_Data.append(t)

    elif ts1 <= tll and ts1 <= ts2:
        t = ts1

    else:
        t = ts2
        ts_Data.append(t)

    tiempo_transcurrido = t - t_evento_anterior
    n_actual = n1 + n2
    tiempo_en_estado[n_actual] = tiempo_en_estado.get(n_actual, 0.0) + tiempo_transcurrido

    area.append(tiempo_transcurrido * n_actual)

    t_evento_anterior = t

    if tll == t:
        n1 = n1 + 1
        tll = t + expovariate(lamda)
        if n1 == 1:
            ts1 = t + expovariate(mu1)

    elif ts1 == t:
        n1 = n1 - 1
        n2 = n2 + 1
        if n1 > 0:
            ts1 = t + expovariate(mu1)
        else:
            ts1 = inf
        if n2 == 1:
            ts2 = t + expovariate(mu2)

    elif ts2 == t:
        n2 = n2 - 1
        count = count + 1
        if n2 > 0:
            ts2 = t + expovariate(mu2)
        else:
            ts2 = inf


for i in range(Num_Pkst):
    d = ts_Data[i] - tll_Data[i]
    r_Data.append(d)

print("Retraso promedio= ", round(mean(r_Data), 4))
print("Número promedio en el sistema total= ", round(sum(area)/t, 4))
print("Probabilidades de estado estacionario (P_n):")
T_final = t
estados_ordenados = sorted(tiempo_en_estado.keys())

for n in estados_ordenados:
    tiempo_en_n = tiempo_en_estado[n]
    probabilidad_n = tiempo_en_n / T_final
    print(f"P({n} clientes en el sistema) = {probabilidad_n:.4f}")

suma_probabilidades = sum(tiempo_en_estado.values()) / T_final
print(f"\nSuma de probabilidades P_n: {suma_probabilidades:.4f}")
```
