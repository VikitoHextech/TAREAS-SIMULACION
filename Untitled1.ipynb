# **Metodo de Monte Carlo CRUDO**
A continuación se estima el valor de la integral:

\begin{equation}
 I = \int_{a=0}^{b = 1} \sqrt{arcatan(x)}~dx
\end{equation}

Sabiendo lo siguiente:

\begin{equation}
  I = \int_{a}^{b} g(x)dx = \int_{a}^{b} w(x)f(x)~dx = E[g(x)] \approx \frac{1}{N} ~ \sum_{N}^{i = 1} g(x_i)
\end{equation}

Teniendo en cuenta que:

\begin{equation}
  w(x) = g(x)(b-a) \\
  f(x) = \frac{1}{b-a} ⇒ U(a,b)
\end{equation}

```python
import random as rdm
import numpy as np
import math as m
import matplotlib.pyplot as plt
from scipy.integrate import quad

def estimarintegral(func, a, b, nmuestras):
    muestras = [rdm.uniform(a, b) for i in range(nmuestras)]  # Generar números random
    sumatoria = sum(func(x) for x in muestras)
    estimado = (b - a) * (1 / nmuestras) * sumatoria
    return estimado

def funcion(x):
    return m.sqrt(m.atan(x))  # g(x)

a = 0
b = 1
valores_n = [10, 20, 30, 40, 50, 750, 1000]
aproximaciones = []
errores = []

valor_real, _ = quad(funcion, a, b)

for nmuestras in valores_n:
    resultado = estimarintegral(funcion, a, b, nmuestras)
    error = abs(valor_real - resultado)
    aproximaciones.append(resultado)
    errores.append(error)

# Graficar la aproximación
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(valores_n, aproximaciones, marker='o', label="Estimación de la Integral")
plt.axhline(y=valor_real, color='r', linestyle='--', label="Valor Real")
plt.xlabel("Número de muestras (N)")
plt.ylabel("Valor de la Integral")
plt.title("Aproximación de la Integral vs. N")
plt.legend()

# Graficar el error
plt.subplot(1, 2, 2)
plt.plot(valores_n, errores, marker='o', color='orange', label="Error Absoluto")
plt.xlabel("Número de muestras (N)")
plt.ylabel("Error")
plt.title("Error vs. N")
plt.legend()

plt.tight_layout()
plt.show()
```
