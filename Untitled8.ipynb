```python
import numpy as np
from scipy.stats import norm

# --- FUNCIONES BASE ---

def black_scholes_call(S, K, T, r, sigma):
    """
    Calcula el precio de un Call Europeo usando Black-Scholes.
    """
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = (S * norm.cdf(d1)) - (K * np.exp(-r * T) * norm.cdf(d2))
    return call_price

def valor_presente_dividendos(dividendos, tiempos, r):
    """
    Calcula el valor presente de una lista de dividendos.
    dividendos: lista de montos [D1, D2...]
    tiempos: lista de tiempos en años [t1, t2...]
    """
    vp = 0
    for div, t in zip(dividendos, tiempos):
        vp += div * np.exp(-r * t)
    return vp

def analizar_ejercicio_americano(dividendos, tiempos, K, r, T_total):
    """
    Analiza si es óptimo ejercer anticipadamente una opción Americana.
    Condición: ¿Es el Dividendo > Intereses ahorrados por pagar el Strike antes?
    D > K * (1 - e^(-r * (T_total - t_div)))
    """
    print("\n--- Análisis de Ejercicio Anticipado (Opción Americana) ---")
    es_optimo_alguno = False

    for i, (div, t) in enumerate(zip(dividendos, tiempos)):
        tiempo_restante = T_total - t
        # Costo de oportunidad de mantener el Strike (Intereses)
        intereses_strike = K * (1 - np.exp(-r * tiempo_restante))

        print(f"Dividendo {i+1} (Mes {t*12:.1f}): ${div:.2f}")
        print(f"  - Tiempo restante hasta vencimiento: {tiempo_restante:.4f} años")
        print(f"  - Intereses ahorrados si se ejerce hoy: ${intereses_strike:.4f}")

        if div > intereses_strike:
            print(f"  >>> RESULTADO: CONVIENE EJERCER anticipadamente en el mes {t*12:.1f}.")
            es_optimo_alguno = True
        else:
            print(f"  >>> RESULTADO: No conviene ejercer (Mejor esperar).")

    if not es_optimo_alguno:
        print("\nConclusión: No es óptimo ejercer en ninguna fecha de dividendo.")

# --- RESOLUCIÓN DEL EJERCICIO 4a ---
def resolver_inciso_a():
    print("="*60)
    print("RESOLUCIÓN INCISO A")
    print("="*60)

    # Datos
    S0 = 105
    K = 100
    r = 0.10      # 10% anual
    sigma = 0.30  # 30% anual
    T = 3/12      # 3 meses en años

    # Dividendo: $2.00 en 1 mes
    div_monto = 2.00
    div_tiempo = 1/12

    # 1. Calcular Valor Presente del Dividendo
    vp_div = div_monto * np.exp(-r * div_tiempo)

    # 2. Ajustar el precio de la acción (S_adj = S0 - VP_dividendos)
    S_ajustado = S0 - vp_div

    # 3. Calcular Black-Scholes con el precio ajustado
    precio_call = black_scholes_call(S_ajustado, K, T, r, sigma)

    print(f"Datos: S0=${S0}, K=${K}, T={T:.2f} años, r={r}, sigma={sigma}")
    print(f"Dividendo: ${div_monto} en el mes 1")
    print(f"VP del Dividendo: ${vp_div:.4f}")
    print(f"Precio Subyacente Ajustado (S'): ${S_ajustado:.4f}")
    print(f"Precio Opción CALL Europea: ${precio_call:.4f} ")

# --- RESOLUCIÓN DEL EJERCICIO 4b y 4c ---
def resolver_inciso_b_c():
    print("\n" + "="*60)
    print("RESOLUCIÓN INCISO B y C")
    print("="*60)

    # Datos
    S0 = 110
    K = 100
    T = 1.0       # 1 año
    r = 0.08      # 8%
    sigma = 0.25  # 25%

    # Dividendos (4 pagos de $1.50)
    montos_divs = [1.50, 1.50, 1.50, 1.50]
    # Tiempos en meses convertidos a años
    meses_divs = [1, 4, 9, 10]
    tiempos_divs = [m/12 for m in meses_divs]

    # --- INCISO B: Valor Call Europeo ---

    # 1. Calcular VP total de los dividendos
    vp_total_divs = valor_presente_dividendos(montos_divs, tiempos_divs, r)

    # 2. Ajustar precio acción
    S_ajustado = S0 - vp_total_divs

    # 3. Calcular B&S
    precio_call = black_scholes_call(S_ajustado, K, T, r, sigma)

    print(f"Datos: S0=${S0}, K=${K}, T={T} año, r={r}, sigma={sigma}")
    print(f"Dividendos: {montos_divs} en los meses {meses_divs}")
    print(f"VP Total Dividendos: ${vp_total_divs:.4f}")
    print(f"Precio Subyacente Ajustado (S'): ${S_ajustado:.4f}")
    print(f"Precio Opción CALL Europea (Inciso B): ${precio_call:.4f} ")

    # --- INCISO C: Análisis Americano ---
    analizar_ejercicio_americano(montos_divs, tiempos_divs, K, r, T)

# --- EJECUCIÓN DEL PROGRAMA ---
if __name__ == "__main__":
    resolver_inciso_a()
    resolver_inciso_b_c()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Asegúrate de tener las funciones anteriores cargadas ---
# (black_scholes_call, etc. del código previo)

def graficar_decision_americana(dividendos, tiempos, K, r, T_total):
    """
    Grafica la comparación entre el Dividendo y el Costo de Oportunidad (Intereses)
    para visualizar cuándo es óptimo ejercer.
    """
    meses = [t * 12 for t in tiempos]
    intereses_costo = []

    # Calcular el umbral de intereses para cada fecha
    for t in tiempos:
        tiempo_restante = T_total - t
        # K * (1 - e^(-r * tiempo_restante))
        costo = K * (1 - np.exp(-r * tiempo_restante))
        intereses_costo.append(costo)

    # Configuración de la gráfica
    fig, ax = plt.subplots(figsize=(10, 6))

    x_pos = np.arange(len(meses))
    width = 0.35

    # Barras de Dividendos
    rects1 = ax.bar(x_pos - width/2, dividendos, width, label='Dividendo a cobrar', color='green', alpha=0.7)
    # Barras de Intereses (Costo)
    rects2 = ax.bar(x_pos + width/2, intereses_costo, width, label='Intereses Ahorrados (Umbral)', color='red', alpha=0.7)

    # Etiquetas y textos
    ax.set_ylabel('Monto ($)')
    ax.set_xlabel('Mes del Pago')
    ax.set_title('Análisis de Ejercicio Temprano (Call Americano)\nSi Barra Verde > Barra Roja -> ¡EJERCER!')
    ax.set_xticks(x_pos)
    ax.set_xticklabels([f'Mes {m:.0f}' for m in meses])
    ax.legend()
    ax.grid(axis='y', linestyle='--', alpha=0.5)

    # Añadir etiquetas de valor encima de las barras
    def autolabel(rects):
        for rect in rects:
            height = rect.get_height()
            ax.annotate(f'${height:.2f}',
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')

    autolabel(rects1)
    autolabel(rects2)

    plt.tight_layout()
    plt.show()

def graficar_sensibilidad_precio(S0, K, T, r, sigma, vp_divs):
    """
    Grafica cómo cambia el precio de la opción (con dividendos)
    ante cambios en el precio de la acción.
    """
    # Generar rango de precios de la acción (+- 20%)
    precios_S = np.linspace(S0 * 0.8, S0 * 1.2, 50)
    precios_Call = []

    for s in precios_S:
        # Ajustamos cada precio por los dividendos constantes (VP)
        s_adj = s - vp_divs
        # Si el precio ajustado es negativo (caso extremo), usamos 0
        s_adj = max(s_adj, 0.01)

        precio = black_scholes_call(s_adj, K, T, r, sigma)
        precios_Call.append(precio)

    plt.figure(figsize=(10, 6))
    plt.plot(precios_S, precios_Call, linewidth=2, label='Precio Call Europeo')

    # Marcar el precio actual
    precio_actual = black_scholes_call(S0 - vp_divs, K, T, r, sigma)
    plt.plot(S0, precio_actual, 'ro', label=f'Precio Actual (${S0}): ${precio_actual:.2f}')

    plt.title(f'Perfil de Valor de la Opción Call (con Dividendos)\nStrike=${K}, T={T} año')
    plt.xlabel('Precio de la Acción ($)')
    plt.ylabel('Precio de la Opción ($)')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

# --- BLOQUE PARA EJECUTAR LAS GRÁFICAS ---
# (Copia esto al final de tu script anterior)

if __name__ == "__main__":
    # Datos del Inciso B/C para graficar
    montos_divs = [1.50, 1.50, 1.50, 1.50]
    meses_divs = [1, 4, 9, 10]
    tiempos_divs = [m/12 for m in meses_divs]
    K_val = 100
    r_val = 0.08
    T_val = 1.0
    S0_val = 110
    sigma_val = 0.25

    # 1. Gráfica de Barras para Ejercicio Americano
    print("Generando gráfica de decisión americana...")
    graficar_decision_americana(montos_divs, tiempos_divs, K_val, r_val, T_val)

    # 2. Gráfica de Sensibilidad (Bonus)
    # Calculamos el VP total de dividendos para este caso
    vp_total = 0
    for d, t in zip(montos_divs, tiempos_divs):
        vp_total += d * np.exp(-r_val * t)

    print("Generando gráfica de sensibilidad...")
    graficar_sensibilidad_precio(S0_val, K_val, T_val, r_val, sigma_val, vp_total)
```

```python
import numpy as np
from scipy.stats import norm

def calcular_black_scholes_total(S, K, T, r, sigma, cantidad):
    """
    Calcula el precio unitario y el costo total de negociar un lote de opciones.
    """
    # 1. Calcular d1 y d2
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    # 2. Calcular Precios Unitarios (Prima por opción)
    # Call = S * N(d1) - K * e^(-rT) * N(d2)
    precio_call = (S * norm.cdf(d1)) - (K * np.exp(-r * T) * norm.cdf(d2))

    # Put = K * e^(-rT) * N(-d2) - S * N(-d1)
    precio_put = (K * np.exp(-r * T) * norm.cdf(-d2)) - (S * norm.cdf(-d1))

    # 3. Calcular Costos Totales de la Posición
    total_call = precio_call * cantidad
    total_put = precio_put * cantidad

    return precio_call, total_call, precio_put, total_put

def resolver_ejercicio_6():
    print("="*60)
    print("RESOLUCIÓN EJERCICIO 6: Valoración de Lote de Opciones")
    print("="*60)

    # --- DATOS DEL PROBLEMA ---
    S0 = 80.00          # Precio del activo
    sigma = 0.40        # Volatilidad (40%)
    cantidad = 100000   # Número de opciones
    r = 0.03            # Tasa libre de riesgo (3%)

    # Tiempo: 6 meses (o 126 días bursátiles estándar)
    # En finanzas T siempre es en años. 6 meses = 0.5 años.
    T = 6 / 12

    # DATO ASUMIDO (No visible en la imagen):
    # Asumimos que es "At The Money" (Strike = Precio Actual)
    K = 80.00

    # --- CÁLCULOS ---
    p_call, t_call, p_put, t_put = calcular_black_scholes_total(S0, K, T, r, sigma, cantidad)

    # --- RESULTADOS ---
    print(f"Datos:")
    print(f"  - Activo (S0): ${S0}")
    print(f"  - Strike (K) [Asumido]: ${K}")
    print(f"  - Volatilidad: {sigma*100}%")
    print(f"  - Tiempo (T): {T} años (6 meses)")
    print(f"  - Tasa (r): {r*100}%")
    print(f"  - Cantidad a negociar: {cantidad:,.0f} opciones")
    print("-" * 60)

    print("RESULTADOS SI ES UNA OPCIÓN DE COMPRA (CALL):")
    print(f"  - Prima unitaria (Precio por opción): ${p_call:.4f}")
    print(f"  > COSTO TOTAL PARA NEGOCIAR EL LOTE:  ${t_call:,.2f}")

    print("\nRESULTADOS SI ES UNA OPCIÓN DE VENTA (PUT):")
    print(f"  - Prima unitaria (Precio por opción): ${p_put:.4f}")
    print(f"  > COSTO TOTAL PARA NEGOCIAR EL LOTE:  ${t_put:,.2f}")

if __name__ == "__main__":
    resolver_ejercicio_6()
```
